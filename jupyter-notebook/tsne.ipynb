```python

import torch
from detectron2.modeling import build_model
from detectron2.config import get_cfg
from detectron2.checkpoint import DetectionCheckpointer
from detectron2.data import MetadataCatalog
import torchvision.transforms as T
from PIL import Image
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import os
```

```python

# 載入配置和模型
cfg = get_cfg()
cfg.merge_from_file("./PTbased-202403/configs/mclass/b2c.yaml")
cfg.MODEL.WEIGHTS = "./output/20240305_moon_b2c/model_0001999.pth"
cfg.MODEL.DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

model = build_model(cfg)
model.eval()
DetectionCheckpointer(model).load(cfg.MODEL.WEIGHTS)
```

```python

# 函數來提取特徵
def extract_features(model, image_path):
    image = Image.open(image_path).convert("RGB")
    transform = T.Compose([
        T.Resize((cfg.INPUT.MIN_SIZE_TEST, cfg.INPUT.MAX_SIZE_TEST)),
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    image = transform(image).unsqueeze(0).to(cfg.MODEL.DEVICE)
    with torch.no_grad():
        features = model.backbone(image)
    # print(features.keys())
        
    return features['res4'].cpu().numpy()

# 假設您有一個圖像列表和相應的標籤
image_folder = '/home/waue0920/pt-moon-202403/data/VOC2007_bddtrain/images/100k/val'
image_paths = [os.path.join(image_folder, img) for img in os.listdir(image_folder) if img.endswith('.jpg')]
labels = [0, 1, ...]  # 假設有8個類別

# 提取特徵
features = np.array([extract_features(model, path) for path in image_paths])


```

```python


# 使用t-SNE降維
tsne = TSNE(n_components=2, random_state=0)
reduced_features = tsne.fit_transform(features.reshape(len(features), -1))

# 繪製t-SNE圖
for i in range(8):
    indices = [j for j, label in enumerate(labels) if label == i]
    plt.scatter(reduced_features[indices, 0], reduced_features[indices, 1], label=f'Class {i}')
plt.legend()
plt.show()
```
